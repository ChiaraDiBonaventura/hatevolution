```python
import pandas as pd
import numpy as np
from scipy.stats import spearmanr

path = # your path here
```

```python
df = pd.read_excel(path+"rankings.xlsx")
```

```python
map_significance2value = {0.80:1.28, 0.90:1.645, 0.95:1.96, 0.99:2.58}

def spearman_ci(x, y, confidence=0.90):
    # Compute Spearman's rank correlation
    rho, _ = spearmanr(x, y)

    # Number of samples
    n = 20

    # Fisher's Z-transformation
    z = 0.5 * np.log((1 + rho) / (1 - rho))

    # Compute the confidence interval in Z-space
    z_crit = map_significance2value[confidence]  # Approximate critical value for alpha % confidence
    z_lower = z - (z_crit / np.sqrt(n - 3))
    z_upper = z + (z_crit / np.sqrt(n - 3))

    # Convert back to Spearman's correlation scale
    rho_lower = (np.exp(2 * z_lower) - 1) / (np.exp(2 * z_lower) + 1)
    rho_upper = (np.exp(2 * z_upper) - 1) / (np.exp(2 * z_upper) + 1)

    return rho, rho_lower, rho_upper
```

```python
ranking_1 = df['Implicit Hate']
ranking_2 = df['Dynabench']

rho, rho_lower, rho_upper = spearman_ci(ranking_1, ranking_2)

print(f"Spearman's rho: {rho:.4f}")
print(f"90% Confidence Interval: ({rho_lower:.3f}, {rho_upper:.3f})")
```
